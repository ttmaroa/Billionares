```python
path = "/content/drive/MyDrive/Billi/Billionaires Statistics Dataset Raw - Copy of Data.csv"
```

```python
import pandas as pd

tt = pd.read_csv(path)
tt
```

```python
tt.columns
```

```python
tt = tt.drop(columns=['firstName', 'lastName', 'birthYear', 'birthMonth', 'birthDay', 'Current'])
```

```python
tt.head()
```

```python
# Looking at the most succesful industries

ind = tt['industries'].value_counts().reset_index()
ind
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='count', y='industries', data=ind)
plt.title('Industries with the Most Billionaires')
plt.xlabel('Number of Billionaires')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

```python
# countries with the most billionares

cc = tt['country'].value_counts().reset_index()
cc
```

```python
top_countries = cc['country'].head(5).tolist()

for country in top_countries:
    print(f"\nIndustries for billionaires in {country}:")
    country_df = tt[tt['country'] == country]
    industries_in_country = country_df['industries'].value_counts().reset_index()
    industries_in_country.columns = ['industry', 'number_of_billionaires']
    display(industries_in_country.head())
```

```python
# Convert 'Year' column to datetime and extract birth year
tt['bd'] = pd.to_datetime(tt['Year'], format='%d/%m/%Y').dt.year


current_year = 2026
tt['age'] = current_year - tt['bd']
```

```python
# no of males and females

tt['gender'].value_counts()
```

```python
# who the top 14 billionares are amd their resepctive industries

top14 = tt.sort_values(by='finalWorth', ascending=False).head(14)

top14[['personName', 'country', 'industries', 'finalWorth']]
```

```python
import plotly.express as px

fig = px.bar(
    top14,
    x='personName',
    y='finalWorth',
    color='industries',
    title="Top 14 Billionaires and Their Industries"
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# looking at the age distribution

plt.figure(figsize=(8,5))
sns.histplot(tt['age'], bins=25)
plt.title("Age Distribution")
plt.show()
```

```python
# Those that are self made and those that aren't

sns.countplot(data=tt, x='selfMade')
plt.title("Self-Made vs Inherited")
plt.show()
```

```python
# who controls more wealth

wealth_by_origin = tt.groupby('selfMade')['finalWorth'].sum()

wealth_by_origin.plot(kind='bar')
plt.title("Total Wealth: Self-Made vs Inherited")
plt.ylabel("Total Final Worth")
plt.show()
```

```python
# what is the life expectancy of countries with the billionares

cs = tt.groupby('country').agg({
    'personName': 'count',
    'life_expectancy_country': 'mean'
}).reset_index()

cs.columns = ['country', 'billionaire_count', 'life_expectancy']
```
